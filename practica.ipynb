```python
#12 - Crea un código que solicite las 3 notas de un estudiante e imprima el promedio de las notas.
nota1 = input("Digite la primera nota: ")
nota2 = input("Digite la segunda nota: ")
nota3 = input("Digite la tercera nota: ")
promedio = (int(nota1) + int(nota2) + int(nota3)) / 3
print(f"El promedio de las notas {nota1}, {nota2} y {nota3} es: {promedio}")
```

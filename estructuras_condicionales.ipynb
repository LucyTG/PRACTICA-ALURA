```python
#Situación:

#Recibiremos el promedio de la nota de los estudiantes y necesitamos de un algoritmo que ejecute 
#el análisis y determine si el estudiante fue Aprobado o Reprobado, mostrando un mensaje del resultado. Para ser aprobado, el promedio necesita ser igual o superior a 7.0.
nota = input("Por favor ingresa una nota del 1 al 10: ")
if float(nota) >= 7:
    print("El estudiante ha aprobado con una nota de:", nota)
if 7 > float(nota) >= 5:
        print("Tienes posibilidad de recuperar la materia.", nota)
else:
    print("El estudiante ha reprobado con una nota de:", nota) 
```

```python
#implementación: del ELIF 
nota = input("Por favor ingresa una nota del 1 al 10: ")
if float(nota) >= 7:
    print("El estudiante ha aprobado con una nota de:", nota)
elif 7 > float(nota) >= 5:
    print("Tienes posibilidad de recuperar la materia.", nota)
else:
    print("El estudiante ha reprobado con una nota de:", nota)
```

```python
#1 - Escribe un programa que pida a la persona usuaria que proporcione dos números y muestre 
# el número más grande.
num1 = float(input("Por favor ingresa el primer número: "))
num2 = float(input("por favor ingresa el segundo número: "))
if num1 > num2:
    print("El número mas grande es ", num1)
elif num2 > num1:
    print("El numero mas grande es ", num2)
else:
    print("Los dos numeros son iguales.")
```

```python
#2 - Escribe un programa que solicite el porcentaje de crecimiento de producción de una empresa e 
# informe si hubo un crecimiento (porcentaje positivo) o una disminución (porcentaje negativo).
porcentaje = float(input("Por favor ingresa el porcentaje de crecimiento de producción de la empresa: "))
if porcentaje > 0:
    print("Hubo un crecimiento del ", porcentaje, "%")
elif porcentaje < 0:
    print("Hubo una disminución del ", porcentaje, "%")
else:
    print("No hubo cambios en la producción de la empresa.")
```

```python
#3 - Escribe un programa que determine si una letra proporcionada por la persona usuaria es una 
# vocal o una consonante.
letra = input("Por favor ingrese una letra: ")
if letra.lower() in ['a', 'e', 'i', 'o', 'u']:
    print("La letra es una vocal.", letra)
else:
    print("La letra es una consonante.", letra)
```

```python
#4 - Escribe un programa que lea valores promedio de precios de un modelo de automóvil durante 3 años 
# consecutivos y muestre el valor más alto y más bajo entre esos tres años.
precio1 = float(input("Por favor ingresa el precio del automóvil en el primer año: "))
precio2 = float(input("Por favor ingresa el precio del automóvil en el segundo año: "))
precio3 = float(input("Por favor ingresa el precio del automóvil en el tercer año: "))
if precio1 > precio2 and precio1 > precio3:
    print("El precio mas alto es del primer año: ", precio1)
elif precio2 > precio1 and precio2 > precio3:
    print("El precio mas alto es del segundo año: ", precio2)
else:
    print("El precio más alto es del tercer año: ", precio3)
if precio1 < precio2 and precio1 < precio3:
    print("El precio más bajo es del primer ano: ", precio1)
elif precio2 < precio1 and precio2 < precio3:
    print("El precio más bajo es del segundo año: ", precio2)
else: 
    print("El precio más bajo es del tercer año: ", precio3)
    
```

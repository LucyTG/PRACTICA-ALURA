```python
#1 - Escribe un programa que solicite dos números enteros e imprima todos 
# los números enteros entre ellos.
num1 = int(input("Por favor ingresa el primer número entero: "))
num2 = int(input("Por favor ingresa el segundo número entero: "))
if num1 < num2:
    for i in range(num1, num2 + 1):
        print(i)
elif num1 > num2:
    for i in range(num2, num1 + 1):
        print(i)
else:
    print("Los dos números son iguales.")
```

```python
#2 - Escribe un programa para calcular cuántos días tomará que la colonia de una bacteria A supere o 
# iguale a la colonia de una bacteria B, basado en tasas de crecimiento del 3% y 1.5%, respectivamente. 
# Supón que la colonia A comienza con 4 elementos y B con 10.
#Colonia.
colonia_A = 4
colonia_B = 10
#Contador de días.
dias = 0
#Tasas de crecimiento.
tasa_A = 0.03
tasa_B = 0.015
#Bucle para calcular los días necesarios.
while colonia_A <= colonia_B:
    colonia_A *= (1 + tasa_A)
    colonia_B *= (1 + tasa_B)
    dias += 1
    # Resultado final
    print("La colonia A iguala o supera a B en", dias, "días")
    
```

```python
#3 - Para procesar una cantidad de 15 datos de evaluaciones de usuarios de un servicio de la empresa, 
# necesitamos verificar si las calificaciones son válidas. Por lo tanto, escribe un programa que 
# recibirá calificaciones del 0 al 5 y verificará si son valores válidos. Si se ingresa una 
# calificación superior a 5 o inferior a 0, se repetirá hasta que el usuario ingrese un valor válido.
for i in range(4):
    nota = float(input(f"Por favor ingresa la calificacion {1+i} entre o y 5: "))
    while nota < 0 or nota > 5:
        print("Calificación inválida. Inténtalo de nuevo.")
        nota = float(input(f"Por favor ingresa la calificacion {1+i} entre o y 5: "))
    print(f"Calificación {1+i} registrada: {nota}")
    
```

```python
#4 - Desarrolla un programa que lea un conjunto indefinido de temperaturas en grados Celsius y calcule su promedio. La lectura debe detenerse al ingresar el valor -273°C.
temperatura = float(input("Ingrese una temperatura en grados Celsius (o -273 para detenerse): "))
suma_temperaturas = 0
contador = 0
while temperatura != -273:
    suma_temperaturas += temperatura
    contador += 1
    temperatura = float(input("Ingrese otra temperatura en grados Celsius (o -273 para detenerse): "))
if contador > 0:
    promedio = suma_temperaturas / contador
    print(f"El promedio de las temperaturas es: {promedio}°C")
else:
    print("No se ingresaron temperaturas válidas.")
```

```python
#5 - Escribe un programa que calcule el factorial de un número entero proporcionado por el usuario.
# Recuerda que el factorial de un número entero es el producto de ese número por todos sus antecesores hasta llegar al número 1. 
# Por ejemplo, el factorial de 5 es 5 x 4 x 3 x 2 x 1 = 120.
numero = int(input("Por favor ingresa un número entero para calcular su factorial: "))
factorial = 1
for i in range(1, numero + 1):
    factorial *= i
print(f"El factorial de {numero} es {factorial}.")
```

```python
#6 - Escribe un programa que genere la tabla de multiplicar de un número entero del 1 al 10, 
# según la elección del usuario. Como ejemplo, para el número 2, la tabla de multiplicar debe 
# mostrarse en el siguiente formato:
numero = int(input("Por favor ingresa un número entero del 1 al 10 para generar su tabla de multiplicar: "))
print(f"Tabla de multiplicar del {numero}:")
for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")
```

```python
#7 - Los números primos tienen diversas aplicaciones en Ciencia de Datos, como en criptografía y 
# seguridad. Un número primo es aquel que es divisible solo por sí mismo y por 1. Por lo tanto, 
# crea un programa que solicite un número entero y determine si es un número primo o no.
numero = int(input("Por favor ingresa un número entero para verificar si es primo: "))
if numero > 1:
    es_primo = True
    for i in range(2, int(numero**0.5) + 1):
        if numero % i == 0:
            es_primo = False
            break
    if es_primo:
        print(f"El número {numero} es primo.")
    else:
        print(f"El número {numero} no es primo.")
else:
    print(f"El número {numero} no es primo.")
```

```python
#8 - Vamos a comprender la distribución de edades de los pensionistas de una empresa de seguros. 
# Escribe un programa que lea las edades de una cantidad no informada de clientes y muestre la distribución 
# en los intervalos [0-25], [26-50], [51-75] y [76-100]. La entrada de datos se detendrá al ingresar un 
# número negativo.
edad = int(input("Por favor ingresa la edad del cliente (o un número negativo para detenerse): "))
rango_0_25 = 0
rango_26_50 = 0
rango_51_75 = 0
rango_76_100 = 0
while edad >= 0:
    if 0 <= edad <= 25:
        rango_0_25 += 1
    elif 26 <= edad <= 50:
        rango_26_50 +=1
    elif 51 <= edad <= 75:
        rango_51_75 += 1
    elif 76 <= edad <= 100:
        rango_76_100 += 1
    edad = int(input("Por favor ingresa la edad del cliente (o un número negativo para detenerse): "))
print("Distribución de edades:")
print(f"[0-25]: {rango_0_25}")
print(f"[26-50]: {rango_26_50}")
print(f"[51-75]: {rango_51_75}")
print(f"[76-100]: {rango_76_100}")
```

```python
#9 - En una elección para la gerencia de una empresa con 20 empleados, hay cuatro candidatos. 
# Escribe un programa que calcule al ganador de la elección. La votación se realizó de la siguiente manera:

# Cada empleado votó por uno de los cuatro candidatos (representados por los números 1, 2, 3 y 4).

# También se contaron los votos nulos (representados por el número 5) y los votos en blanco 
# (representados por el número 6).

#Al final de la votación, el programa debe mostrar el número total de votos para cada candidato, 
# los votos nulos y los votos en blanco. Además, debe calcular y mostrar el porcentaje de votos nulos 
# con respecto al total de votos y el porcentaje de votos en blanco con respecto al total de votos.
# Inicializamos las variables contadoras
votos_candidato1 = 0
votos_candidato2 = 0
votos_candidato3 = 0
votos_candidato4 = 0
votos_nulos = 0
votos_blanco = 0

# Inicio del bucle para leer los votos
for i in range(0, 20):
    voto = int(input('Ingresa tu voto: '))
    if voto == 1:
        votos_candidato1 += 1
    elif voto == 2:
        votos_candidato2 += 1
    elif voto == 3:
        votos_candidato3 += 1
    elif voto == 4:
        votos_candidato4 += 1
    elif voto == 5:
        votos_nulos += 1
    elif voto == 6:
        votos_blanco += 1
    else:
        print("Voto inválido.")

print(f'Votos candidato 1: {votos_candidato1}')
print(f'Votos candidato 2: {votos_candidato2}')
print(f'Votos candidato 3: {votos_candidato3}')
print(f'Votos candidato 4: {votos_candidato4}')
print(f'Votos nulos: {votos_nulos}')
print(f'Votos en blanco: {votos_blanco}')
print(f'Porcentaje de votos nulos: {(votos_nulos / 20 * 100)}')
print(f'Porcentaje de votos en blanco: {(votos_blanco / 20 * 100)}')
```
